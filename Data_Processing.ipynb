```python
df_industry = df_industry.drop(['RIC','COMPANY NAME','NAME'],axis=1)
```

```python
industry = list(df_industry['ICB INDUSTRY NAME'])
symbol = df_industry['Symbol']
name = list(df_industry['Name'])
```

File dữ liệu Stock

```python
df = pd.ExcelFile('Philippine-Price-Market-Cap.xlsx')
# Đọc sheet 1
df_1 = pd.read_excel(df,'Price',index_col=0,parse_dates=True)
# Đọc sheet 2
df_2 = pd.read_excel(df,'Market Cap',index_col=0,parse_dates=True)
```

```python
# Sheet1 transportation
price_sheet = df_1.T

price_sheet.insert(0, "Industry", industry)

price_sheet = price_sheet.T

# Sheet2 transportation
marketcap_sheet = df_2.T

marketcap_sheet.insert(0, "Industry", industry)

marketcap_sheet = marketcap_sheet.T
```

```python
# Viết hàm xóa bỏ các ngành Error

def delete_error(df):
    for i in df.columns:
        if 'ERROR' in i:
            df = df.drop(i,axis=1)
    return df

price_sheet = delete_error(price_sheet)
marketcap_sheet = delete_error(marketcap_sheet)
```

```python
# Kiểm tra dữ liệu NaN
print(price_sheet.iloc[0].isnull().sum())

print(marketcap_sheet.iloc[0].isnull().sum())
```

```python
# Xóa dữ liệu NaN trong ngành sheet 1
x = np.where(price_sheet.iloc[0].isnull())

price_sheet = price_sheet.drop(price_sheet.columns[x], axis=1) 

# Xóa dữ liệu NaN trong ngành sheet 2
y = np.where(marketcap_sheet.iloc[0].isnull())

marketcap_sheet = marketcap_sheet.drop(marketcap_sheet.columns[y], axis=1) 
```

```python
# Kiểm tra dữ liệu NaN
print(price_sheet.iloc[0].isnull().sum())

print(marketcap_sheet.iloc[0].isnull().sum())
```

```python
# Phương thức xóa dữ liệu NaN của giá cổ phiếu
def drop_columns_with_na(df):
    for i in df.columns:
        if df[i].isnull().any():
            df=df.drop(i,axis=1)
    return df

price_sheet = drop_columns_with_na(price_sheet)

marketcap_sheet = drop_columns_with_na(marketcap_sheet)



print('Tổng missing data sheet 1:',price_sheet.isnull().sum().sum())
print('Tổng missing data sheet 2:',marketcap_sheet.isnull().sum().sum())
```

```python
# Xuất file đã clean data
price_sheet.to_excel("data_cleaned_final_sheet_1.xlsx")
marketcap_sheet.to_excel("data_cleaned_final_sheet_2.xlsx")
```
