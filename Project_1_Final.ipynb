```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import datetime
import matplotlib
import seaborn as sns
```

```python
sheet_1 = pd.read_excel('data_cleaned_final_sheet_1.xlsx')
sheet_1
```

```python
#Lược bỏ '00:00:00' ở dữ liệu về ngày
sheet_1['Name']=sheet_1['Name'].astype('str')
sheet_1['Name']=sheet_1['Name'].apply(lambda x: x.split(' ')[0])
sheet_1.set_index('Name',inplace=True)
sheet_1
```

```python
# Thay các số 0 trong DF thành Mean()
mean_storing = [] # lưu mean theo index
temp_df = sheet_1.iloc[1:] # bỏ dòng đầu vì là chuỗi
count = 0
indexCol_drop = []
for i in temp_df.columns:
    if temp_df[i].mean(skipna=True) != 0:
        mean_storing.append(temp_df[i].mean(skipna=True))
    else:
        indexCol_drop.append(count)
    count+=1
sheet_1.drop(sheet_1.columns[indexCol_drop],axis = 1,inplace=True)

counts = 0
for i in sheet_1.columns:
    sheet_1[i].replace(to_replace=0,value=mean_storing[counts],inplace=True)
    counts+=1
```

***Câu 1***

```python
#Liệt kê 10 cổ phiếu có SSL cao nhất so với ngày đầu tháng
cau_1 = sheet_1.T
cau_1 = cau_1[['2021-09-01','2021-09-24']]
cau_1['SSL']= (cau_1['2021-09-24']-cau_1['2021-09-01'])/(cau_1['2021-09-01'])
cau_1=cau_1.sort_values(by='SSL',ascending=False).head(10)
cau_1
```

```python
#Vẽ biểu đồ
fig, ax = plt.subplots(figsize= (18,6))

fig_2 = sns.barplot(x= cau_1.index,y='SSL',data= cau_1,palette='Greens_d',saturation =0.5)

ax.set_xticklabels(cau_1.index,rotation = 90, fontsize= 13)

ax.set_title('10 cổ phiếu có tỷ suất sinh lợi cao nhất so với ngày đầu tháng (01/09/2021)',loc= 'center',fontsize= 15);
```

***Câu 5***

```python
capital = sheet_2.loc[['Industry','2021-09-24']]
capital = capital.T
```

```python
# Xuất tổng vốn hóa ngành

capital_industry = capital.groupby('Industry')['2021-09-24'].sum()
capital_industry = capital_industry.sort_values(ascending=False)
```

```python
#Quy về %
sum_capital_industry = capital_industry.sum()
pct_capital = (capital_industry/sum_capital_industry)*100
print('Cơ cấu vốn thị trường     Đơn vị (%)')
print(pct_capital)
```

```python
#Tạo ra một dải màu cho biểu đồ
cmap = matplotlib.cm.get_cmap('viridis')

rgba = cmap([1-0.05*i for i in range(len(pct_capital))])
```

```python
#Vẽ biểu đồ
fig, ax = plt.subplots(figsize=(13,8),ncols=2, sharey=True)
fig.tight_layout()
ax[0].barh(pct_capital.index.tolist(), (pct_capital.values/2).tolist(), align='center', color=list(rgba), zorder=10)
ax[1].barh(pct_capital.index, (pct_capital.values/2).tolist(), align='center', color=list(rgba), zorder=10)
ax[0].invert_xaxis() 
plt.gca().invert_yaxis()
ax[0].set(yticks=pct_capital.index.tolist(), yticklabels=pct_capital.index.tolist())
ax[0].yaxis.tick_left()
ax[0].tick_params(axis='y', colors='black')
plt.subplots_adjust(wspace=0, top=0.85, bottom=0.1, left=0.18, right=0.95)
for spine in ax[0].spines:
    ax[0].spines[spine].set_visible(False)
for spine in ax[1].spines:
    ax[1].spines[spine].set_visible(False)
ax[0].get_xaxis().set_visible(False)
ax[1].get_xaxis().set_visible(False)
ax[1].get_yaxis().set_visible(False)
totals = []
for i in ax[1].patches:
    totals.append(i.get_width())
total = sum(totals)
for i in ax[1].patches:
    ax[1].text(i.get_width()+.5, i.get_y()+0.6, \
            str(round((i.get_width()/total)*100, 2))+'%', fontsize=15,
color='black')
fig.suptitle('BỨC TRANH TOÀN CẢNH VỀ TỔNG GIÁ TRỊ VỐN HÓA CỦA CÁC NGÀNH TRÊN SÀN CHỨNG KHOÁN PHILIPPINES',fontsize=18)
plt.show()
```
